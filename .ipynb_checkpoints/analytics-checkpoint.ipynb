```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

% matplotlib inline
```

```python
df = pd.read_csv('07-2018 data.csv', delimiter=',')
df.head()
```

```python
df[df['nome'].str.contains('Marques')]
```

```python
#Escolaridade dos servidores
df['escolaridade'].value_counts().plot('pie')
```

```python
#Cargos
df['cargo'].value_counts().plot('pie')
```

```python
#Cargos por escolaridade
scholarity = 'MESTRADO'
df[df['escolaridade'] == scholarity]['cargo'].value_counts().plot('pie')
```

```python
#Funções por escolaridade
scholarity = 'DOUTORADO'
df[df['escolaridade'] == scholarity]['funcao'].value_counts().plot('pie')
```

```python
#Media de salários (Bruto) das escolaridades (descendente)
df.groupby(['escolaridade']).mean().sort_values('total_bruto', ascending=False)#.apply(pd.to_numeric, errors='none').mean()
```

```python
#Maiores salários (Bruto)
df.sort_values('total_bruto', ascending=False).head(10).drop(['nome'], axis=1)
```
